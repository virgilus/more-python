## ğŸ“¦ Package managers

Python is great but managing packages can sometimes be a challenge.

### ğŸ“¦ What is a package?

A package is a collection of modules that can be installed and used in your Python projects. Packages can include libraries, frameworks, and tools that provide additional functionality to your code.

### ğŸ”§ What is a package manager?

A package manager is a tool that helps you install, update, and manage packages in your Python environment. It simplifies the process of finding and installing packages, as well as managing dependencies between packages.

### ğŸ¤· What are the options?

You have many options for managing packages in Python, the most popular ones include:

- `pip`: The default package manager for Python, included with most Python installations.
- `conda`: A package and environment management system that can manage packages for Python and other languages.
- `uv`: A fast and modern package manager written in rust for Python that focuses on speed and simplicity.
- `pdm`: A modern Python package and dependency manager that aims to simplify the management of Python projects.
- `poetry`: A dependency management and packaging tool for Python that simplifies the process of managing dependencies and building packages.
- `pipenv`: A tool that combines `pip` and `virtualenv` to provide a unified way to manage dependencies and virtual environments.
- `setuptools`: A package development and distribution tool that includes a package manager for installing and managing packages.

### ğŸ `pip`

`pip`, which stands for "Pip Installs Packages," (a recursive acronym) is the most widely used package manager for Python. It allows you to install packages from the Python Package Index (PyPI), which is the official third-party software repository for Python, and other package repositories.

It will automatically handle dependencies for you, ensuring that all required packages are installed alongside the package you want.

All the packages installed via `pip` are stored in a directory called `site-packages`, which is located within your Python installation or virtual environment.

The basic commands for `pip` include:

- `pip install package_name`: Installs a package. When no version is specified, the latest version is installed.

If you want to make sure which python environment you are using with `pip`, you can run:

```bash
python -m pip install package_name
```

The `-m` flag tells Python to run the `pip` module as a script, ensuring that the package is installed in the correct environment.

- `pip uninstall package_name`: Uninstalls a package.
- `pip list`: Lists all installed packages.
- `pip freeze`: Outputs a list of installed packages in a format that can be used to recreate the environment. However, unless you want to recreate the exact same environment for production, prefer using `pip list`.
- 
- `pip show package_name`: Displays information about a specific package.
  
- `pip install -r requirements.txt`: Installs packages listed in a requirements file.

Don't put all your dependencies in a single file. You should specify versions using comparison operators to avoid compatibility issues. For example:

```
pandas >=2.0.0
numpy >=1.21.0
requests <=2.25.0
```

You can also use pip-compile from `pip-tools` to generate a `requirements.txt` file with pinned versions from a `requirements.in` file. For example, let's say your requirements.in file looks like this:

```
# requirements.in
pandas>=2.1
requests>=2.31
```

Then you can run:
```bash
pip-compile requirements.in
```

This will generate a `requirements.txt` file with the exact versions of the packages and their dependencies:

```# requirements.txt
pandas==2.1.0
numpy==1.24.3
python-dateutil==2.8.2
pytz==2023.3
requests==2.31.0
certifi==2023.5.7
charset-normalizer==3.1.0
idna==3.4
urllib3==1.26.15
```

- `pip install --upgrade package_name`: Upgrades an installed package to the latest version.
- `pip check`: Checks for broken dependencies in the installed packages.

- `pip install -e .`: Installs a package in editable mode, allowing you to make changes to the source code and have them reflected in the installed package. Because when you use `pip install package_name`, pip copies the package files to the `site-packages` directory. Any changes made to the original source code will not be reflected in the installed package unless you reinstall it.

### ğŸ Conda

`conda` is another popular package manager that can manage packages for Python and other languages. It is often used in data science and scientific computing because it can handle complex dependencies and provides precompiled binaries for many packages.

Why should you use conda over pip?

- Environment management: `conda` has built-in support for creating and managing virtual environments, making it easy to isolate dependencies for different projects.
- Cross-language support: `conda` can manage packages for multiple programming languages, not just Python.
- Precompiled binaries: `conda` provides precompiled binaries for many packages, which can save time and effort when installing packages with complex dependencies.
- Better dependency resolution: `conda` uses a more sophisticated dependency resolution algorithm than `pip`, which can help avoid conflicts between packages.

If you work inside a conda environment, prefer using `conda` to install packages when possible. If a package is not available via `conda`, you can use `pip` as a fallback.

- `conda install package_name`: Installs a package.
- `conda remove package_name`: Uninstalls a package.
- `conda list`: Lists all installed packages.
- `conda update package_name`: Upgrades an installed package to the latest version.

### âš¡ uv

`uv` is a fast and modern package manager written in Rust for Python that focuses on speed and simplicity. It aims to provide a better user experience than `pip` by being faster and more reliable. Its main features include:

The basic commands for `uv` when it comes to install packages are very similar to `pip` and `conda`:

- `uv install package_name`: Installs a package.
- `uv remove package_name`: Uninstalls a package.
- `uv list`: Lists all installed packages.
- `uv update package_name`: Upgrades an installed package to the latest version.

The only downside is that `uv` is relatively new and may not have as many packages available as `pip` or `conda`.

### ğŸ“œ Poetry

Poetry is a dependency management and packaging tool for Python that simplifies the process of managing dependencies and building packages. It provides a simple and intuitive interface for managing dependencies, creating virtual environments, and building packages.

You can start with poetry by creating a new project:

```bash
poetry init
```

This command will guide you through the process of creating a new `pyproject.toml` file, which is used to manage your project's dependencies and settings.

And now if you want to install packages, you can use the following commands:

- `poetry add package_name`: Installs a package and adds it to the `pyproject.toml` file.
- `poetry remove package_name`: Uninstalls a package and removes it from the `pyproject.toml` file.
- `poetry show`: Lists all installed packages.
- `poetry update package_name`: Upgrades an installed package to the latest version.

Then if you want to recreate the environment on another machine, you can simply run:

```bash
poetry install
```

This command will read the `pyproject.toml` file and install all the required dependencies in a new virtual environment.

### ğŸ—ï¸ PDM

`PDM` (Python Development Master) is a modern Python package and dependency manager that aims to simplify the management of Python projects. It uses the `pyproject.toml` file to manage dependencies and project settings, similar to Poetry.

It might replace `poetry` in the future as it is faster and more lightweight.

- Simplified dependency management: PDM makes it easy to add, remove, and update dependencies using simple commands.
- Virtual environment management: PDM can automatically create and manage virtual environments for your projects.
- PEP 582 support: PDM supports PEP 582, which allows you to manage dependencies without the need for virtual environments.
- Build and publish packages: PDM provides tools for building and publishing Python packages to PyPI or other package repositories.

### ğŸ§ª Pipenv

`Pipenv` is a tool that combines `pip` and `virtualenv` to provide a unified way to manage dependencies and virtual environments. It simplifies the process of creating and managing virtual environments, as well as installing and updating packages.

But its development has slowed down in recent years, and other tools like `poetry` and `pdm` have gained popularity as alternatives.

### âš™ï¸ Setuptools

`Setuptools` is a package development and distribution tool that includes a package manager for installing and managing packages. It provides a simple way to create and distribute Python packages, as well as manage dependencies. But nowadays, it is mostly used for packaging and distributing Python projects rather than managing dependencies.

### ğŸ“ TL;DR

- Use `pip` for most cases.
- `conda` if you need complex dependencies or work in data science and use conda environments.
- consider `pdm` (and maybe `poetry`) for modern dependency management and packaging.

# ğŸŒ Virtual environments

When coding in Python, it's a good practice to use virtual environments to manage dependencies for different projects. This helps avoid conflicts between packages and ensures that each project has its own isolated environment.

You have many options for creating virtual environments in Python, including:

- `pyenv`: A popular tool for managing multiple Python versions and creating virtual environments.
- `venv`: A built-in module in Python 3 for creating lightweight virtual environments.
- `virtualenv`: A popular third-party tool for creating virtual environments.
- `conda`: A package and environment management system that can create virtual environments for Python and other languages.
- `uvenv`: A fast and modern tool for creating and managing virtual environments.

## ğŸ Conda

`conda` is a popular choice for managing virtual environments, especially in data science and scientific computing. You can create a new conda environment with the following command:

```bash
conda create --name myenv python=3.11
```

Then you can activate the environment with:

```bash
conda activate myenv
```

To deactivate the environment, simply run:

```bash
conda deactivate
```

Good practices are:
- create a new conda environment for each project to avoid dependency conflicts.
- Try to not install anything in the base environment.
- use `conda env export > environment.yml` to export the environment configuration to a YAML file, which can be used to recreate the environment on another machine with `conda env create -f environment.yml`.

This yaml file would look like this:

```yaml
name: myenv
channels:
  - defaults
dependencies:
  - python=3.11
  - numpy=1.24.3
  - pandas=2.1.0
  - pip:
    - requests==2.31.0
```
